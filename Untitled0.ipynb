# Etapa 06) Tratando os outliers
Dica: Vamos considerar estes valores como erro.Visualize os outliers e

```python
#Removendo os valores com erros:
df_filter = df_new[(df_new['Quantity'] <= 10000) & (df_new['UnitPrice'] <= 5000)]

df_filter.describe()
```

# Etapa 07) Crie uma coluna adicional

```python
#Criar uma coluna nova com o pre√ßo total da compra

df_filter ['TotalCompra'] = df_filter['Quantity'] * df_filter['UnitPrice']

df_filter.head()
```

# Etapa 08) √öltima data

```python
#Data da √∫ltima compra
df_ultima_compra = df_filter['InvoiceDate'].max()

df_ultima_compra
```

# Etapa 09) Plotando gr√°ficos


*   Top 10 pa√≠ses com maior valor em vendas
*   Top 10 produtos mais vendidos
*   Valor de venda total por m√™s
*   Valor de venda total por m√™s e por pa√≠s (considere apenas os top 10)









```python
#Valore de vendas por pa√≠s:
sales_country = df_filter.groupby('Country')['TotalCompra'].sum().sort_values(ascending= False)

#Selecionar Top 10:
top_country = sales_country.head(10)

#Plotar gr√°fico:
plt.figure(figsize=(10,6))
top_country.plot(kind='bar',color= 'blue')
plt.title('Top 10 pa√≠ses com maior valor em vendas')
plt.xlabel('Pa√≠s')
plt.ylabel('Valor total em vendas')
plt.show()
```

```python
#top 10 produtos mais vendidos:
sales_products = df_filter.groupby('Description')['Quantity'].sum().sort_values(ascending= False)

#Selecionar Top 10:
top_products = sales_products.head(10)

#Plotar gr√°fico:
plt.figure(figsize=(10,6))
top_products.plot(kind='bar',color= 'blue')
plt.title('Top 10 produtos mais vendidos')
plt.xlabel('Produto')
plt.ylabel('Quantidade')
plt.show()
```

```python
#Calcular valor de vendas por m√™s
sales_per_month = df_filter.groupby(df_filter['InvoiceDate'].dt.to_period('M'))['TotalCompra'].sum()

#Plotar gr√°fico
plt.figure(figsize=(12,7))
sales_per_month.plot(kind='line',color= 'blue')
plt.title('Valor de vendas por m√™s')
plt.xlabel('M√™s')
plt.ylabel('Valor')
plt.show()
```

```python
# Selecionar os top 10 pa√≠ses com maior valor em vendas
top_10_countries = df_filter.groupby('Country')['TotalCompra'].sum().nlargest(10)

# Filtrar o DataFrame para incluir apenas os top 10 pa√≠ses
df_top_10_countries = df_filter[df_filter['Country'].isin(top_10_countries.index)]

# Criar a tabela din√¢mica para o valor de vendas por m√™s e por pa√≠s
sales_per_month_country = df_top_10_countries.groupby([df_top_10_countries['InvoiceDate'].dt.to_period('M'), 'Country'])['TotalCompra'].sum().unstack()

# Plotar o gr√°fico
plt.figure(figsize=(12, 8))
sales_per_month_country.plot(kind='line', marker='o')
plt.title('Valor de Venda Total por M√™s e por Pa√≠s (Top 10)')
plt.xlabel('M√™s')
plt.ylabel('Valor Total de Vendas')
plt.legend(title='Pa√≠s', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

# Etapa 10) C√°lculo do RFM
üí° Dica: Agrupe os dados por cliente e pedido/compra (InvoiceNo) e obtenha a
data e o pre√ßo total do pedido.
Com isso, agrupe novamente apenas por cliente e calcule o RFM, onde:
- R √© a rec√™ncia, diferen√ßa em dias da √∫ltima compra do cliente e da √∫ltima
compra dispon√≠vel no conjunto de dados, que calcularam previamente.
- F √© a frequ√™ncia, ou seja, a quantidade de compras feitas pelo cliente;
- M √© o ticket m√©dio, ou seja, a m√©dia das compras feitas pelo cliente.

```python

#Calculo da recorr√™ncia (R)
recency =df_filter.groupby('CustomerID')['InvoiceDate'].max() #data da √∫ltima compra de cada cliente
recency = (df_ultima_compra - recency).dt.days

#Calcular a frequencia de compras de cada cliente (F)
frequency = df_filter.groupby('CustomerID')['InvoiceNo'].nunique()

#Calcular o total de gastos por cliente
total_client = df_filter.groupby('CustomerID')['TotalCompra'].sum()

#Calcular o ticket m√©dio (M):
average_spent = total_client / frequency

df_rfm = pd.DataFrame({'Recency':recency,'Frequency':frequency,'Monetary':average_spent})

df_rfm.head()

```
